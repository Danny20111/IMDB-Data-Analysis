## Classify Movies Based on Ratings [Excellent, Good and Average]

```python
data.columns
```

```python
def rating(rating):
    if rating>7.0:
        return 'Excellent'
    elif rating >=6.0:
        return 'Good'
    else:
        return 'Average'
```

```python
data['rating_cat']=data['Rating'].apply(rating)
```

## Find Unique Values from Genre

```python
data.columns
```

```python
data['Genre']
```

```python
list1 = []
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
one_d = []
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
uni_list=[]
for i in one_d:
    if i not in uni_list:
        uni_list.append(i)
```
